## 2.1.1. Getting Started

```
import torch
```

torch를 불러옵니다.

```
x = torch.arange(12)
y = torch.arange(12, dtype = torch.float32)
```

torch가 포함하고 있는 arange라는 함수를 활용하여 0부터 11까지의 숫자를 생성하여 x라는 tensor에 저장을 하였습니다. y라는 tensor는 dtype을 정의해 주었습니다. ***어떤 차이가 있을까요?***

```
x
```

```
y
```

변수명만 입력하고 실행을 하게되면 변수의 내용을 출력합니다.

```
x.shape
```

```
x.size()
```

```
x.type()
```

```
y.type()
```

tensor는 attribute 및 function 들을 가지고 있어 .을 붙이고 그것들을 보고나 활용할 수 있습니다. 

```
x.shape
x.size()
x.type()
```

여러 개의 변수를 연속으로 입력 후에 실행을 하면 마지막 변수만을 보여줍니다.

```
print(x.shape)
print(x.size())
print(x.type())
```

print함수를 이용하면 각각을 모두 볼 수 있습니다.

```
x = x.reshape((3,4))
print(x)
print(x.shape)
```

tensor가 가지고 있는 reshape라는 함수를 이용하여 tensor의 shape를 변경한 것 입니다. 총 12개의 숫자를 가지고 있기 때문에 3x4 matrix로 변환을 할 수 있었습니다.

```
x = x.reshape((3,5))
```

3x5로 변환을 하라고 하면 에러가 납니다. 15개의 숫자가 있어야 하기 때문입니다.

```
x = x.reshape((3,-1))
x
```

-1을 사용하는 것은 매우 유용합니다. 3만 지정하고 -1을 넣으면 값이 유효한 경우 스스로 크기를 맞추어 줍니다.

```
x = x.reshape((-1,4))
x
```

```
x = x.reshape((-1,5))
x
```

유효하지 않은 숫자를 넣은 경우에는 에러가 납니다.

```
x = torch.zeros((2,3,4))
print(x)
print(x.shape)
```

모든 원소가 0인 3차원 matrix를 2x3x4의 크기로 만들어 보았습니다. 아래는 모든 원소가 1인 경우로 만드는 것입니다.

```
w = torch.ones((2,3,4))
print(w)
print(w.shape)
```

```
y = torch.tensor([[1,2,3,4],[2,3,4,1],[3,4,1,2]])
y
```

위와 같이 직접 모든 값을 입력해서 생성할 수도 있습니다.

```
n = torch.randn(2,3,4)
n
```

특정한 확률분포를 따르는 숫자들을 생성할 필요가 자주 있습니다. randn은 평균이 0이고 표준편차가 1인 normal distribution으로 숫자들을 생성합니다.

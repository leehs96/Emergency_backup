## 2.1.2. Operations

```
import torch
```

```
x = torch.tensor([1,2,4,8])
y = torch.tensor([2,2,2,2])
print('x=',x)
print('x+y=',x+y)
print('x-y=',x-y)
print('x*y=',x*y)
print('x/y=',x/y)
```

위의 연산들은 element-wise operation들입니다. 대응하는 요소들끼리 연산을 하게 됩니다. 그런데 나누기의 결과값에 이상한 부분이 있습니다.

```
x = torch.tensor([1,2,4,8])
y = torch.tensor([2,2,2,2])
print('x/y=',x/y)
x = torch.tensor([1.,2,4,8])
y = torch.tensor([2.,2,2,2])
print('x/y=',x/y)
x = torch.tensor([1,2,4,8],dtype=torch.float32)
y = torch.tensor([2,2,2,2],dtype=torch.float32)
print('x/y=',x/y)
```

프로그래밍을 하면서 항상 주의해 주어야 할 부분 중 하나는 변수의 타입입니다.

```
torch.exp(x)
```

```
torch.exp([-1.0,0.0,1.0])
```

torch의 함수들은 대부분 input이 tensor의 형태로 존재해야 하는 것에 주의합시다.

```
torch.exp(torch.tensor([-1.0,0.0,1.0]))
```

```
x = torch.arange(12, dtype=torch.float32).reshape((3,4))
y = torch.tensor([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]], dtype=torch.float32)
print('x=',x)
print('y=',y)
print('y.t()=',y.t())
print('xy=',torch.mm(x, y.t()))
```

matrix multiplication은 mm이라는 함수를 통해 이루어 집니다. 이를 위해 y는 .t()를 이용해 transpose를 해주었습니다.

```
x = torch.arange(4, dtype=torch.float32).reshape((2,2))
y = torch.tensor([[2, 1], [1, 2]], dtype=torch.float32)
print('x=',x)
print('y=',y)
```

x라는 tensor와 y라는 tensor를 합치는 것을 아래 두 경우를 통해 확인해 봅시다. dim은 dimension을 나타냅니다.

```
torch.cat((x,y),dim=0)
```

```
torch.cat((x,y),dim=1)
```

```
x = torch.arange(4, dtype=torch.float32).reshape((2,2))
y = torch.tensor([[2, 1], [1, 2]], dtype=torch.float32)
print('x=',x)
print('y=',y)
```

```
x == y
```

두 tensor의 요소들이 일치하는 경우가 궁금할 수 있습니다. 정답을 맞추어야 되는 경우 이런 operation이 필요할 것입니다. 사이즈가 일치하는 경우 ==을 사용해주면 됩니다.

```
x.sum()
```

.sum()은 x의 모든 요소들을 더한 결과를 출력합니다.

```
x > y
```

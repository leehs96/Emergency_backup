## 2.1.3. Broadcasting Mechanism

```
import torch
```

```
a = torch.tensor([[0.],[1],[2]])
b = torch.tensor([0.,1])
print('a =',a)
print('b =',b)
```

a는 3x1이고 b는 1x2입니다. 요소의 개수가 다르므로 일반적으로는 더하기 연산을 수행할 수 없습니다.

```
a + b
```

그런데 pytorch에서는 broadcasting이라는 과정을 통해 이러한 경우에도 더하기 연산을 수행합니다.

```
print(a-b)
print(a*b)
print(a/b)
```

## 2.1.5. Saving Memory

```
x = torch.tensor([1.,2,3])
y = torch.tensor([3.,2,1])
print(id(x))
print(id(y))
```

Python의 id라는 함수는 변수가 저장되어 있는 메모리의 위치를 나타냅니다.

```
before = id(y)
print(before)
y = y + x
print(id(y))
id(y) == before
```

원래 y가 저장되어 있는 메모리의 위치를 before에 기록하고, y = y + x를 수행한 후 새로운 y가 저장되어 있는 메모리의 위치를 비교해보니 다른 것을 알 수 있습니다. 이는 새로운 y를 위해 새로운 메모리의 위치를 배정한 것입니다. 파라미터가 계속 변하고 이전의 값을 기억할 필요가 없는 경우에는 그 때마다 새로운 메모리 공간을 할당하는 것은 비효율적입니다. 

```
before = id(y)
print(before)
y[:] = y + x
print(id(y))
id(y) == before
```

 [:]을 사용하면 새로운 메모리를 할당하지 않고 원래의 위치에 값을 바꾸어 저장합니다. 

```
before = id(y)
print(before)
torch.add(y,x,out=y)
print(id(y))
id(y) == before
```

torch.add를 이용하는 경우에도 새로운 메모리를 할당하지 않고 원래의 위치에 값을 바꾸어 저장합니다.

```
before = id(y)
print(before)
y += x
print(id(y))
id(y) == before
```

+=연산을 이용해도 마찬가지의 결과를 얻을 수 있습니다.

# 2.4. Calculus

## 2.4.1. Derivatives and Differntiation

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x):
  return 3*x**2-4*x
def numerical_lim(f,x,h):
  return (f(x+h)-f(x))/h
```

f(x)와 f'(x)를 정의하였습니다. 

```
h = 0.1
for i in range(5):
  print('h=%.5f, numerical limit=%.5f' % (h, numerical_lim(f,1,h)))
  h *= 0.1
```

h의 값을 점점 작게함에 따라서 x=1에서의 f'(x)의 값이 달라지는 것을 확인할 수 있습니다. 

```
print(f(1))
print(np.arange(.1,5,0.5))
print(f(np.arange(.1,5,0.5)))
print(numerical_lim(f,0.001,np.arange(.1,5,0.5)))
```

```
x = np.arange(.1,5,0.25)
fx = f(x)
dfx = numerical_lim(f,0.001,x)
plt.plot(x,fx,label="f(x)")
plt.plot(x,dfx,label="f'(x)")
plt.xlabel("x")
plt.ylabel("f(x) & f'(x)")
plt.legend()
plt.show()
```

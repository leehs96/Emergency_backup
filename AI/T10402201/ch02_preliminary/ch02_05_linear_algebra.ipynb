# 2.3. Linear Algebra

```
import torch
```

## 2.3.1. Scalars

```
x = torch.tensor([3.])
y = torch.tensor([2.])
print('x + y = ',x+y, ', x * y = ',x*y)
print('x / y = ', x/y, ', x ** y = ',torch.pow(x,y))
```

scalar는 하나의 값을 tensor에 넣어주면 됩니다.

```
print(type(x))
print(type(x.numpy()))
print(type(x.item()))
```

x는 pytorch tensor입니다. 
.numpy()를 쓰면 numpy.ndarray으로 값을 반환합니다.
.item()을 쓰면 python float으로 값을 반환합니다.

## 2.3.2. Vectors

```
x = torch.arange(4)
print(x)
y = torch.arange(4,dtype=torch.float32)
print(y)

print('y[3] = ', y[3])
```

vector는 여러 값을 하나의 축을 가지도록 tensor에 넣어주면 됩니다.

### 2.3.2.1. Length, Dimensionality, and Shape

```
x = torch.arange(4)
print(len(x))
print(x.shape)
```

len은 python에 기본적으로 있는 함수입니다. .shape는 pytorch에서 제공합니다.

## 2.3.3. Matrices

```
A = torch.arange(20, dtype = torch.float32).reshape((5,4))
A
```

20개의 요소를 가지는 vector를 생성한 후에, 5x4의 matrix로 변환하였습니다.

```
print(A.t())
```

.t()를 이용해 matrix's transpose를 얻을 수 있습니다.

```
B = torch.tensor([[1,2,3],[2,0,4],[3,4,5]])
print('B = ',B)
BT = B.t()
print('BT = ',BT)
```

```
B == BT
```

matrix B는 B's transpose와 동일합니다. 이런 matrix를 symmetric matrix라고 합니다.

## 2.3.5. Basic Properties of Tensor Artithmetic

```
A = torch.arange(12).reshape(3,4)
B = A.clone()
print('A = ',A)
print('A + B =',A+B)
```

B는 A를 복제하여 A+B를 출력하였습니다.

```
a = 2
X = torch.arange(12).reshape(3,4)
print('a + X = ', a+X)
print('a * X = ', a*X)
print('shape of a * X : ', (a*X).shape)
```

## 2.3.6. Reduction


```
x = torch.arange(4)
print('x = ',x)
print('x.sum() = ',x.sum())
```

```
A = torch.arange(4).reshape((2,2))
print('A = ',A)
print('A.sum() = ',A.sum())
```

.sum()은 모든 요소들을 합한 값을 반환합니다.

```
A = torch.arange(6,dtype=torch.float32).reshape((2,3))
print('A = ',A)
A_sum_axis0 = A.sum(axis=0)
print('A_sum_axis0 = ',A_sum_axis0)
A_sum_axis1 = A.sum(axis=1)
print('A_sum_axis1 = ',A_sum_axis1)
```

특정 축을 기준으로 하여 합한 값을 반환할 수 있습니다.

```
print(A)
print(A.sum(axis=[0,1]))
```

```
print(torch.mean(A))
print(torch.sum(A))
print(torch.numel(A))
print(torch.div(torch.sum(A),torch.numel(A)))
```

.mean()을 이용해 평균값을 취할 수 있고, .sum()과 .numel()을 이용해 평균값을 계산할 수도 있습니다.

## 2.3.7. Dot Products

```
x = torch.arange(4, dtype = torch.float32)
y = torch.ones(4, dtype = torch.float32)
print(x)
print(y)
print(torch.sum(x*y))
print(torch.dot(x,y))
```

dot product는 .dot()을 통해 계산할 수 있습니다.

## 2.3.8. Matrix-vector products

```
A = torch.arange(6,dtype=torch.float32).reshape((2,3))
x = torch.arange(3,dtype=torch.float32)
print(A)
print(x)
print(torch.mv(A,x))
```

matrix와 vector의 곱은 .mv()를 이용합니다.

## 2.3.9. Matrix-matrix multiplication

```
B = torch.arange(6,dtype=torch.float32).reshape((3,2))
print(A)
print(B)
print(torch.mm(A,B))
```

matrix와 matrix의 곱은 .mm()을 이용합니다.

## 2.3.10. Norms

```
x = -1*torch.arange(4,dtype=torch.float32)
print(x)
print(torch.abs(x))
print(torch.norm(x))
print(torch.sum(torch.abs(x)))
```

l2 norm은 .norm()을 통해 계산합니다. l1 norm은 .abs()로 절대값을 취한 후에 .sum()을 해주면 됩니다.
